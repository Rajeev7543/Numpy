# Aggregating function in numpy

```python
a= np.array([20,30,40,50])
print(np.sum(a))
print(np.min(a))
print(np.max(a))
print(np.cumsum(a))
print(np.mean(a))
print(np.cumprod(a))
```

```python
a= [120,130,140,150,170,190]
b=[11,13,14,12,19,18]
price=np.array(a)
quantity = np.array(b)
print(price,"\n",quantity)
print()
c=(np.cumproduct([price,quantity],axis=0))
print(c[1].sum())
```

# Statistical function

```python
import statistics as stats
baked_food = [200,400,150,120,300]

a= np.array(baked_food)
print(np.mean(baked_food))
print(np.median(baked_food))
print(stats.mode(baked_food))
print(np.std(baked_food)) # standar daviation ke square ko variance bolte hai 
```

```python
# -1 represent inversely proportional relationship
# 1 represent proportional relationship
# 0 means no relationship
R=[20,30,10,50,40,80]
birth=[100,200,300,120,130,140]
print(np.corrcoef([R,birth])) # ek ke badhne par dusra badhe 
```
